```python
def format_opcodes(opcode_str):
    opcodes = opcode_str.split()
    formatted = ""
    i = 0

    # Опкоды, после которых нужен дополнительный перенос строки
    newline_opcodes = {'JUMPDEST', 'REVERT', 'JUMPI', 'JUMP', 'STOP', 'RETURN', 'MSTORE'}

    index = 0

    while i < len(opcodes):
        opcode = opcodes[i]

        if opcode.startswith("PUSH0"):
            formatted += f"{hex(index)} {opcode}\n"
            index += 1
            i += 1
        elif opcode.startswith("PUSH"):
            operand = opcodes[i + 1]
            formatted += f"{hex(index)} {int(opcode.split('H')[1])} {opcode} {operand}\n"
            index += int(opcode.split('H')[1])
            i += 2
        else:
            formatted += f"{hex(index)} {opcode}\n"
            index += 1
            i += 1

        if opcode in newline_opcodes:
            formatted += "\n"

    return formatted.strip()
```
